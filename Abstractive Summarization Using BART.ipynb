### **BART Fine-Tuning: Using Transformers**

```python
import torch
from torch.nn import functional as F
from torch import nn
import pytorch_lightning as pl
# https://www.pytorchlightning.ai/

from transformers import BartForConditionalGeneration, BartTokenizer #These are used for loading the BART model and tokenizing text.
from sklearn.model_selection import train_test_split
import pandas as pd

from transformers import (
    AdamW,
    get_linear_schedule_with_warmup
)
from torch.utils.data import DataLoader
```

```python
#These classes provide a well-structured framework for loading and preprocessing data, making it ready for fine-tuning the BART
#model using PyTorch Lightning. It demonstrates how to prepare the data in a format that can be directly consumed by the model
#during training.

class Dataset(torch.utils.data.Dataset):#custom implementation of a PyTorch Dataset, used to organize and preprocess the data for the BART model.

#initializes the dataset with text, summaries, tokenizer, and maximum sequence lengths for the source and summaries.
    def __init__(self, texts, summaries, tokenizer, source_len, summ_len):
        self.texts = texts
        self.summaries = summaries
        self.tokenizer = tokenizer
        self.source_len  = source_len
        self.summ_len = summ_len

    def __len__(self):#method returns the number of samples in the dataset.
        return len(self.summaries) - 1
#method preprocesses and tokenizes a single data sample, including the source text and the summary.
#It returns input IDs and attention masks for both source and target sequences.
    def __getitem__(self, index):
        text = ' '.join(str(self.texts[index]).split())

        summary = ' '.join(str(self.summaries[index]).split())

        # Article text pre-processing
        source = self.tokenizer.batch_encode_plus([text],
                                                  max_length= self.source_len,
                                                  pad_to_max_length=True,
                                                  return_tensors='pt')
        # Summary Target pre-processing
        target = self.tokenizer.batch_encode_plus([summary],
                                                  max_length= self.summ_len,
                                                  pad_to_max_length=True,
                                                  return_tensors='pt')

        return (
            source['input_ids'].squeeze(), 
            source['attention_mask'].squeeze(), 
            target['input_ids'].squeeze(),
            target['attention_mask'].squeeze()
        )

# PyTorch Lightning DataModule, responsible for loading and preprocessing data using the Dataset class.
class BARTDataLoader(pl.LightningDataModule):
    '''Pytorch Ligtning Model Dataloader class for BART'''

#method initializes the data loader with necessary parameters like tokenizer, text length, summarized length, file path, corpus size, columns' names, train split size, and batch size.
    def __init__(self, tokenizer, text_len, summarized_len, file_path,
                 corpus_size, columns_name, train_split_size, batch_size):
        super().__init__()
        self.tokenizer = tokenizer
        self.text_len = text_len
        self.summarized_len = summarized_len
        self.input_text_length = summarized_len
        self.file_path = file_path
        self.nrows = corpus_size
        self.columns = columns_name
        self.train_split_size = train_split_size
        self.batch_size = batch_size

#method reads and preprocesses the data from a CSV file, adding the "summarize:" prefix to each summary.
    def prepare_data(self):
        data = pd.read_csv(self.file_path, nrows=self.nrows,
                               encoding='latin-1')
        data = data[self.columns]
        data.iloc[:, 1] = 'summarize: ' + data.iloc[:, 1]
        self.text =  list(data.iloc[:, 0].values)
        self.summary =  list(data.iloc[:, 1].values)

    def setup(self, stage=None): #method splits the data into training and validation sets.
        X_train , y_train, X_val, y_val = train_test_split(
            self.text, self.summary
        )

        self.train_dataset = (X_train, y_train) 
        self.val_dataset = (X_val, y_val)

# method creates and returns a DataLoader for the training data using the Dataset class.
    def train_dataloader(self):
        train_data = Dataset(texts=self.train_dataset[0],
                             summaries=self.train_dataset[1],
                             tokenizer=self.tokenizer,
                             source_len=self.text_len,
                             summ_len=self.summarized_len)
        return DataLoader(train_data, self.batch_size)

#method creates and returns a DataLoader for the validation data using the Dataset class.
    def val_dataloader(self):
        val_dataset = Dataset(texts=self.val_dataset[0],
                             summaries=self.val_dataset[1],
                             tokenizer=self.tokenizer,
                             source_len=self.text_len,
                             summ_len=self.summarized_len)
        return DataLoader(val_dataset, self.batch_size)
```

```python
# defines a PyTorch Lightning module tailored for fine-tuning a BART-based model on abstractive summarization.
# It encapsulates the training, validation, and optimization processes
class AbstractiveSummarizationBARTFineTuning(pl.core.LightningModule): #class is derived from pl.core.LightningModule.
    

    def __init__(self, model, tokenizer):
        super().__init__()
        self.model = model #the BART model for fine-tuning
        self.tokenizer = tokenizer #used for preprocessing

# method defines the forward pass of the model. It takes input IDs, attention masks, decoder input IDs, and optional arguments
# the BART model is called using the provided inputs.
# method returns the outputs of the BART model, including the loss
    def forward(self, input_ids, attention_mask, decoder_input_ids,
                decoder_attention_mask=None, lm_labels=None):
        
        outputs = self.model.forward(
            input_ids=input_ids,
            attention_mask=attention_mask,
            decoder_input_ids=decoder_input_ids,
            labels=decoder_input_ids
        )

        return outputs
# method restructures the batch data into the required format for the model.
# It extracts and renames the relevant elements from the batch input, such as input IDs, attention masks, decoder input IDs, etc.

    def preprocess_batch(self, batch):
        """Reformatting batch"""

        input_ids, source_attention_mask, decoder_input_ids, \
        decoder_attention_mask = batch

        y = decoder_input_ids
        decoder_ids = decoder_input_ids
        source_ids = input_ids
        source_mask = source_attention_mask

        return source_ids, source_mask, decoder_ids, decoder_attention_mask, decoder_attention_mask
    
# method is called during each training iteration.
# It preprocesses the batch using the preprocess_batch method and then calls the forward method.
# The loss from the model's outputs is returned

    def training_step(self, batch, batch_idx):
        """Model Training Step computes """

        input_ids, source_attention_mask, decoder_input_ids, \
        decoder_attention_mask, lm_labels = self.preprocess_batch(batch)

        outputs = self.forward(input_ids=input_ids, attention_mask=source_attention_mask,
                               decoder_input_ids=decoder_input_ids,
                               decoder_attention_mask=decoder_attention_mask,
                               lm_labels=lm_labels 
                       )
        loss = outputs.loss

        return loss
    
# method is called during each validation iteration.
# Similar to the training step, it preprocesses the batch, calls the forward method, and returns the loss.

    def validation_step(self, batch, batch_idx):
        """"""

        input_ids, source_attention_mask, decoder_input_ids, \
        decoder_attention_mask, lm_labels = self.preprocess_batch(batch)

        outputs = self.forward(input_ids=input_ids, attention_mask=source_attention_mask,
                               decoder_input_ids=decoder_input_ids,
                               decoder_attention_mask=decoder_attention_mask,
                               lm_labels=lm_labels 
                       )
        loss = outputs.loss

        return loss
    
# The training_epoch_end method is called at the end of each training epoch.
# It computes the average loss for the epoch and logs it using self.log.



    def training_epoch_end(self, outputs):
        avg_loss = torch.stack([x["loss"] for x in outputs]).mean()
        self.log('Epoch', self.trainer.current_epoch)
        self.log('avg_epoch_loss', {'train': avg_loss})

# The val_epoch_end method is called at the end of each validation epoch.
# It computes the average validation loss and logs it. 

    def val_epoch_end(self, loss):
        avg_loss = torch.stack([x["loss"] for x in loss]).mean()
        self.log('avg_epoch_loss', {'Val': avg_loss})

# method defines the optimizer for the model's parameters.
# An AdamW optimizer is created using the model's parameters.
# The optimizer is returned as a list.

    def configure_optimizers(self):

        model = self.model
        optimizer = AdamW(model.parameters())
        self.opt = optimizer

        return [optimizer]
```

```python
#sets up the environment for loading and preprocessing the dataset, splitting it into training and validation sets,
# and creating a data loader that can be used for fine-tuning the BART model. It demonstrates the steps required to prepare data
# for training a summarization model using PyTorch Lightning.
# Tokenizer
# Upload the curated_data_subset.csv if using colab or change the path to local file
model_ = BartForConditionalGeneration.from_pretrained("facebook/bart-base")
tokenizer = BartTokenizer.from_pretrained("facebook/bart-base")

# Dataloader
dataloader = BARTDataLoader(tokenizer=tokenizer, text_len=512,
                            summarized_len=150,
                            file_path='curated_data_subset.csv',
                            corpus_size=50, columns_name=['article_content','summary'],
                            train_split_size=0.8, batch_size=2)
# Read and pre-process data
dataloader.prepare_data()

# Train-test Split
dataloader.setup()
```

```python
# Main Model class
# An instance of the class is created, passing the preloaded BART model (model_) and tokenizer (tokenizer) as arguments.
model = AbstractiveSummarizationBARTFineTuning(model=model_, tokenizer=tokenizer)
```

```python
#lass from PyTorch Lightning is instantiated to manage the training process.
trainer = pl.Trainer(check_val_every_n_epoch=1, max_epochs=5)
#check_val_every_n_epoch parameter is set to 1, which means validation will be performed after every epoch.

# Fit model
trainer.fit(model, dataloader) #method is called to start the training process.
```

# **BART Abstractive Summarization: Using Pre Trained Model**

```python
from transformers import BartTokenizer, BartForConditionalGeneration, BartConfig
```

```python
def summarize_article(article):
    # Load BART model and tokenizer
    model_name = 'facebook/bart-large-cnn' #corresponds to a pre-trained BART model specifically fine-tuned for CNN/Daily Mail data.
    tokenizer = BartTokenizer.from_pretrained(model_name)
    model = BartForConditionalGeneration.from_pretrained(model_name)

    # Tokenize and encode the article
    inputs = tokenizer.encode(article, return_tensors='pt',
max_length=1024, truncation=True)

    # Generate summary
    summary_ids = model.generate(inputs, num_beams=4, max_length=150,
early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary

# Example usage
article = """
My friends are cool but they eat too many carbs.
"""

summary = summarize_article(article)
print("Summary:")
print(summary)
```
